```python
import pandas as pd
from plotnine import *
```

```python
mean_temps = pd.read_csv("../../data/mean_temps.csv")
mean_temps.head()
```

```python
( ggplot(mean_temps, aes(x = 'month', y = 'location', fill = 'mean')) + 
  geom_tile() +
  scale_fill_cmap(limits = [0.15, 0.98], name = "temperature (F)")
)
```
