```python
import pandas as pd
from plotnine import *
```

```python
mean_temps = pd.read_csv("../../data/mean_temps.csv")
mean_temps.head()
```

```python
mean_temps.dtypes
```

```python
( ggplot(mean_temps, aes(x = 'month', y = 'location', fill = 'mean')) + 
     geom_tile(aes(width=0.95,height=0.95)) +
     scale_fill_cmap(cmap_name='inferno', name = "temperature (F)") +
     scale_x_discrete(limits =['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']) +
     scale_y_discrete(name = "", limits =['Death Valley', 'Houston', 'San Diego', 'Chicago']) +
     theme(figure_size=(10, 3),
          panel_background=element_rect(fill='white', alpha=.3))
)
```
