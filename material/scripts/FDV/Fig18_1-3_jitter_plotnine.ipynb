```python
import pandas as pd
from plotnine import *
```

```python
mpg = pd.read_csv("../../data/mpg.csv")
mpg.head()
```

```python
p_mpg_transp = ( ggplot(mpg, aes(y = 'cty', x = 'displ', color = 'drv', fill = 'drv')) +
  geom_point(size = 3, alpha=0.4)  + 
  ylab("fuel economy (mpg)") +
  xlab("displacement (l)")  +
  scale_color_manual(values=["#202020", "#E69F00", "#56B4E9"], 
                     name="drive train",
                     breaks=["f", "r", "4"],
                     labels=["FWD", "RWD", "4WD"]) +
  scale_fill_manual(values=["#20202080", "#E69F0080", "#56B4E980"], 
                     name="drive train",
                     breaks=["f", "r", "4"],
                     labels=["FWD", "RWD", "4WD"]) +
  theme_classic() +
  theme(legend_position = (.7, .8),
        legend_direction = "vertical"
  #       plot_margin = [3, 7, 3, 1.5])
       )
               )
p_mpg_transp
```

```python
p_mpg_jitter = ( ggplot(mpg, aes(y = 'cty', x = 'displ', color = 'drv', fill = 'drv')) +
  geom_point(size = 3,
             position = position_jitter(width = 0.01 * (mpg["displ"].max() - mpg["displ"].min()),
                                        height = 0.01 * (mpg["cty"].max() - mpg["cty"].min()),
                                        random_state = 7384)) + 
  ylab("fuel economy (mpg)") +
  xlab("displacement (l)") +
  theme_classic() +
  scale_color_manual(values=["#202020", "#E69F00", "#56B4E9"], 
                     name="drive train",
                     breaks=["f", "r", "4"],
                     labels=["FWD", "RWD", "4WD"]) +
  scale_fill_manual(values=["#20202080", "#E69F0080", "#56B4E980"], 
                     name="drive train",
                     breaks=["f", "r", "4"],
                     labels=["FWD", "RWD", "4WD"]) +
  # theme_dviz_open() +
  theme(legend_position = (.7, .8),
        legend_direction = "vertical"
        # plot.margin = margin(3, 7, 3, 1.5))
                
               )
               )

p_mpg_jitter
```
