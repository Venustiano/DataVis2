```r
library(data.table)
library(ggplot2)
```

```r
dt <- fread("data/penguins.csv")
```

```r
dodge <- position_dodge(width = 0.9)
ggplot2::ggplot(dt, ggplot2::aes(x = species,y = body_mass_g,fill = sex)) +
  ggplot2::geom_violin(position = dodge)+
	ggplot2::theme_bw() +
  ggplot2::theme(plot.title = ggplot2::element_text(hjust = 0.5)) +
ggplot2::geom_boxplot(position = dodge,colour="#636363", width = 0.1,alpha=0.2)
```

```r
ggplot2::ggplot(dt, ggplot2::aes(x = species,y = body_mass_g,group = sex,colour=sex)) +
  ggplot2::geom_violin()+
	ggplot2::theme_bw() +
  ggplot2::theme(plot.title = ggplot2::element_text(hjust = 0.5)) +
ggplot2::geom_boxplot(colour="#636363", width = 0.1,alpha=0.2)
```
