```python
import pandas as pd
from plotnine import *
```

```python
cows_dens = pd.read_csv("cows_dens.csv")
cows_dens
```

```python
(
    ggplot(cows_dens, aes(x = 'butterfat', y = 'density', color = 'breed', fill = 'breed')) + 
  geom_density(stat = "identity") +
  scale_color_manual(
    values = ["#56B4E9", "#E69F00", "#D55E00", "#009E73"],
    breaks = ["Ayrshire", "Guernsey", "Holstein-Friesian", "Jersey"]
  ) +
  scale_fill_manual(
    values = ["#56B4E950", "#E69F0050", "#D55E0050", "#009E7350"],
    breaks = ["Ayrshire", "Guernsey", "Holstein-Friesian", "Jersey"]
  ) +
  scale_x_continuous(
    expand = [0, 0],
    name = "butterfat contents (%)"
  ) +
  scale_y_continuous(limits = [0, 1.99], expand = [0, 0])+
    theme_bw()
)
```
