## Library

```r
library(ggfortify)
```

## Data

```r
wtable <- read.csv("../../data/exergamewf.csv")
head(wtable)
```

## PCA

```r
tpca <- prcomp(wtable[,5:15],scale=TRUE)
#print(str(balance.pca))
tpca
```

## Plot

```r
fig <- autoplot(tpca, data = wtable, colour = 'older',loadings = TRUE, loadings.colour = 'blue',
         loadings.label = TRUE, loadings.label.size = 5) +
    theme_bw()
fig
```
